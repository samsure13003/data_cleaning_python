```python
def clean_countries(s):
    # Split by comma
    countries = s.split(',')
    clean_list = []
    for c in countries:
        c = c.strip()                 # remove spaces
        c = re.sub(r'en:', '', c)     # remove en: prefix
        c = c.lower()                 # lowercase for uniformity
    return list(set(clean_list))       # remove duplicates per row

data_cleaned = country_data_series.apply(clean_countries)
```

```python
country_map = {
    'maroc': 'morocco',
    'marruecos': 'morocco',
    'argelia': 'algeria',
    'belgica': 'belgium',
    'suiza': 'switzerland',
    'francia': 'france',
    'alemania': 'germany',
    'reunion': 'r√©union',
    'irlande': 'ireland',
}

def normalize_countries(lst):
    return [country_map.get(c, c) for c in lst]

data_normalized = data_cleaned.apply(normalize_countries)
```
